```python
# Function for finding the top correlations
```

```python
def get_redundant_pairs(df):
    '''Get diagonal and lower triangular pairs of correlation matrix'''
    pairs_to_drop = set()
    cols = df.columns
    for i in range(0, df.shape[1]):
        for j in range(0, i+1):
            pairs_to_drop.add((cols[i], cols[j]))
    return pairs_to_drop
```

```python
def get_top_abs_correlations(df, n=5):
    au_corr = df.corr().abs().unstack()
    labels_to_drop = get_redundant_pairs(df)
    au_corr = au_corr.drop(labels=labels_to_drop).sort_values(ascending=False)
    return au_corr[0:n]
```

```python
# Creating function for checking VIF
from statsmodels.stats.outliers_influence import variance_inflation_factor
def calc_vif(X):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)
```

# **Training and Evaluation of Model-1**

```python
# Writing the function for cross validation, hyperparameter tuning and evaluation
def rf_model(X_train, y_train, X_test, y_test, parameters, n_fold):
  '''
  This function is for cross validation, training and evaluation of the model

  INPUT: X_train = independent training set as numpy ndarray, y_train= target training set as pandas series, 
  X_test= independent testing set as numpy ndarray, y_test= target testing set as pandas series, 
  parameters= hyperparameters dictionary, n_fold= integers 2,3,5 etc.
  
  OUTPUT: prints the performance metric for train and test
  
  '''
  try:
    # Importing libraries
    from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
    from sklearn.ensemble import RandomForestRegressor
    from sklearn.metrics import r2_score
    from sklearn.metrics import mean_squared_error

    # Model cross validation and hyperparameter tuning
    my_model = RandomForestRegressor()
    my_model_cv = RandomizedSearchCV(my_model, param_distributions=parameters, scoring='r2', cv=n_fold, random_state=0, n_jobs=-1)  # cv=3 means 3 fold cross validation
    my_model_cv.fit(X_train,y_train)

    # Model training (considered earlier best parameters)
    my_model_final = RandomForestRegressor(**my_model_cv.best_params_)
    my_model_final.fit(X_train,y_train)

    # Checking model performance for train set
    y_train_preds = my_model_final.predict(X_train)
    train_rmse=int(math.sqrt(mean_squared_error(10**y_train, 10**y_train_preds)))
    print("Train RMSE:",train_rmse)
    train_r2=round(r2_score(10**y_train, 10**y_train_preds),2)
    print("Train R2:",train_r2)

    # Checking model performance for test set
    y_test_preds = my_model_final.predict(X_test)
    test_rmse=int(math.sqrt(mean_squared_error(10**y_test, 10**y_test_preds)))
    print("Test RMSE:",test_rmse)
    test_r2=round(r2_score(10**y_test, 10**y_test_preds),2)
    print("Test R2:",test_r2)

  except:
    print("Sorry ! Please check your input!")
  
  return print(f"Best parameters:\n{my_model_cv.best_params_}")
```

```python
parameters = {'n_estimators':[int(x) for x in np.linspace(50,200,num=15)], 
              'max_depth':[int(x) for x in np.linspace(5,30,num=6)],
              'max_features':['auto','sqrt'],
              'min_samples_split':[2,5,10],
              'min_samples_leaf':[1,2,5]}
```

```python
rf_model(X_train, y_train, X_test, y_test, parameters, 3)
```

# **Training  and Evaluation of Model-2**

```python
# Writing the function for cross validation, hyperparameter tuning and evaluation
def ridge_model(X_train, y_train, X_test, y_test, parameters, n_fold):
  '''
  This function is for cross validation, training and evaluation of the model

  INPUT: X_train = independent training set as numpy ndarray, y_train= target training set as pandas series, 
  X_test= independent testing set as numpy ndarray, y_test= target testing set as pandas series, 
  parameters= hyperparameters dictionary, n_fold= integers 2,3,5 etc.
  
  OUTPUT: prints the performance metric for train and test
  
  '''
  try:
    # Importing libraries
    from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
    from sklearn.linear_model import Ridge
    from sklearn.metrics import r2_score
    from sklearn.metrics import mean_squared_error

    # Model cross validation and hyperparameter tuning
    my_model=Ridge()
    my_model_cv = RandomizedSearchCV(my_model, param_distributions=parameters, scoring='r2', cv=n_fold, random_state=0)
    my_model_cv.fit(X_train,y_train)

    # Model training (considered earlier best parameters)
    my_model_final = Ridge(**my_model_cv.best_params_)
    my_model_final.fit(X_train,y_train)

    # Checking model performance for train set
    y_train_preds = my_model_final.predict(X_train)
    train_rmse=int(math.sqrt(mean_squared_error(10**y_train, 10**y_train_preds)))
    print("Train RMSE:",train_rmse)
    train_r2=round(r2_score(10**y_train, 10**y_train_preds),2)
    print("Train R2:",train_r2)

    # Checking model performance for test set
    y_test_preds = my_model_final.predict(X_test)
    test_rmse=int(math.sqrt(mean_squared_error(10**y_test, 10**y_test_preds)))
    print("Test RMSE:",test_rmse)
    test_r2=round(r2_score(10**y_test, 10**y_test_preds),2)
    print("Test R2:",test_r2)

  except:
    print("Sorry ! Please check your input!")
  
  return print(f"Best parameters:\n{my_model_cv.best_params_}")
```

```python
parameters = {'alpha':[int(x) for x in np.linspace(10,200,num=20)], 
              'max_iter':[int(x) for x in np.linspace(10,200,num=20)], 'tol':[0.1,0.2],'fit_intercept':[True]}
```

```python
# Writing the function for cross validation, hyperparameter tuning and evaluation
def catboost_model(X_train, y_train, X_test, y_test, parameters, n_fold):
  '''
  This function is for cross validation, training and evaluation of the model

  INPUT: X_train = independent training set as numpy ndarray, y_train= target training set as pandas series, 
  X_test= independent testing set as numpy ndarray, y_test= target testing set as pandas series, 
  parameters= hyperparameters dictionary, n_fold= integers 2,3,5 etc.
  
  OUTPUT: prints the performance metric for train and test
  
  '''
  try:
    # Importing libraries
    from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
    from catboost import CatBoostRegressor, FeaturesData, Pool
    from sklearn.metrics import r2_score
    from sklearn.metrics import mean_squared_error

    # Model cross validation and hyperparameter tuning
    my_model = CatBoostRegressor()
    my_model_cv = RandomizedSearchCV(estimator=my_model, param_distributions = parameters, cv = n_fold, n_jobs=-1)
    my_model_cv.fit(X_train, y_train)

    # Model training
    my_model_final = CatBoostRegressor(**my_model_cv.best_params_)
    my_model_final.fit(X_train,y_train)

    # Checking model performance for train set
    y_train_preds = my_model_final.predict(X_train)
    train_rmse=int(math.sqrt(mean_squared_error(10**y_train, 10**y_train_preds)))
    print("Train RMSE:",train_rmse)
    train_r2=round(r2_score(10**y_train, 10**y_train_preds),2)
    print("Train R2:",train_r2)

    # Checking model performance for test set
    y_test_preds = my_model_final.predict(X_test)
    test_rmse=int(math.sqrt(mean_squared_error(10**y_test, 10**y_test_preds)))
    print("Test RMSE:",test_rmse)
    test_r2=round(r2_score(10**y_test, 10**y_test_preds),2)
    print("Test R2:",test_r2)

  except:
    print("Sorry ! Please check your input!")
  
  return print(f"Best parameters:\n{my_model_cv.best_params_}")
```

```python
parameters = {'depth' : [6,8,10],
              'learning_rate' : [0.01, 0.05, 0.1],
              'iterations'    : [30, 50, 100]}
```

```python
# Finding the performance metrics
catboost_model(X_train, y_train, X_test, y_test, parameters, 3)
```
